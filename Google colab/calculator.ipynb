```python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        raise ZeroDivisionError("Cannot divide by zero")
    return x / y

print(add(5, 10))
print(subtract(10, 5))
print(multiply(4, 5))
print(divide(20, 5))
```
