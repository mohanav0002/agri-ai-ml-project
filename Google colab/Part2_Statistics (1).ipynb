## Section 2.1: Descriptive Statistics

### Key Concepts
- **Mean:** Average value (sum / count)
- **Median:** Middle value when sorted (robust to outliers)
- **Mode:** Most frequent value
- **Std Dev:** Measure of spread from mean
- **Variance:** Square of standard deviation
- **Range:** Max - Min

```python
# Generate normal distribution
np.random.seed(42)
normal_data = np.random.normal(loc=100, scale=15, size=10000)

print("NORMAL DISTRIBUTION")
print("=" * 70)
print(f"Generated from: mean=100, std_dev=15, n=10000")
print()

# Analyze generated data
print("Generated Statistics:")
print(f"  Mean: {np.mean(normal_data):.2f}")
print(f"  Std Dev: {np.std(normal_data):.2f}")
print(f"  Min: {np.min(normal_data):.2f}")
print(f"  Max: {np.max(normal_data):.2f}")
print()

# Empirical rule
within_1std = np.sum((normal_data > 85) & (normal_data < 115)) / len(normal_data) * 100
within_2std = np.sum((normal_data > 70) & (normal_data < 130)) / len(normal_data) * 100
within_3std = np.sum((normal_data > 55) & (normal_data < 145)) / len(normal_data) * 100

print("Empirical Rule (68-95-99.7 rule):")
print(f"  Within 1 std dev [85, 115]: {within_1std:.1f}% (theoretical: 68%)")
print(f"  Within 2 std dev [70, 130]: {within_2std:.1f}% (theoretical: 95%)")
print(f"  Within 3 std dev [55, 145]: {within_3std:.1f}% (theoretical: 99.7%)")
print()

# Plot
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(normal_data, bins=50, density=True, alpha=0.7, color='blue')
plt.axvline(np.mean(normal_data), color='red', linestyle='--', label=f'Mean={np.mean(normal_data):.1f}')
plt.axvline(np.mean(normal_data) + np.std(normal_data), color='orange', linestyle='--', label=f'±1 Std Dev')
plt.axvline(np.mean(normal_data) - np.std(normal_data), color='orange', linestyle='--')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 2: Different distributions
plt.subplot(1, 2, 2)
skewed_data = np.random.exponential(scale=30, size=10000)
plt.hist(skewed_data, bins=50, density=True, alpha=0.7, color='green')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Right-Skewed Distribution (e.g., Income)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("INTERPRETATION")
print("=" * 70)
print("Normal distribution is symmetric (bell curve)")
print("Real data often has skew (like income: most people earn less)")
print("Different distributions need different handling!")
```

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

# Sample dataset - student test scores
scores = np.array([65, 70, 75, 80, 85, 90, 95])

print("DESCRIPTIVE STATISTICS")
print("=" * 70)
print(f"Data: {scores}")
print()

# Mean
mean = np.mean(scores)
print("1. MEAN (Average)")
print("-" * 70)
print(f"Formula: sum(values) / count")
print(f"Calculation: ({' + '.join(map(str, scores))}) / {len(scores)}")
print(f"           = {np.sum(scores)} / {len(scores)}")
print(f"Result: {mean}")
print(f"Interpretation: On average, the score is {mean}")
print()

# Median
median = np.median(scores)
print("2. MEDIAN (Middle Value)")
print("-" * 70)
print(f"Sorted: {np.sort(scores)}")
print(f"Middle position: {(len(scores) + 1) // 2}")
print(f"Result: {median}")
print(f"Interpretation: Half the scores are below {median}, half above")
print(f"Note: Median = Mean when data is symmetric")
print()

# Standard Deviation
std = np.std(scores)
print("3. STANDARD DEVIATION (Measure of Spread)")
print("-" * 70)
print(f"Formula: √(sum((x - mean)²) / n)")
print(f"Result: {std:.2f}")
print(f"Interpretation:")
print(f"  - Typical score varies by ±{std:.2f} points from the mean")
print(f"  - Small std dev = consistent scores")
print(f"  - Large std dev = varied scores")
print()

# Variance
variance = np.var(scores)
print("4. VARIANCE (Square of Std Dev)")
print("-" * 70)
print(f"Formula: sum((x - mean)²) / n")
print(f"Result: {variance:.2f}")
print(f"Note: Variance = Std Dev²")
print(f"      {variance:.2f} = {std:.2f}²")
print()

# Min and Max
print("5. RANGE (Max - Min)")
print("-" * 70)
print(f"Min: {np.min(scores)}")
print(f"Max: {np.max(scores)}")
print(f"Range: {np.max(scores) - np.min(scores)}")
print()

# Percentiles
print("6. PERCENTILES (Dividing data into parts)")
print("-" * 70)
q1 = np.percentile(scores, 25)
q2 = np.percentile(scores, 50)  # Same as median
q3 = np.percentile(scores, 75)
print(f"Q1 (25th percentile): {q1}  → 25% of scores below this")
print(f"Q2 (50th percentile): {q2}  → 50% of scores below this (=median)")
print(f"Q3 (75th percentile): {q3}  → 75% of scores below this")
```

## Section 2.3: Correlation - Finding Relationships

### Correlation Coefficient
- Ranges from -1 to +1
- **+1:** Perfect positive relationship
- **0:** No relationship
- **-1:** Perfect negative relationship

### Uses
- Feature selection (keep features correlated with target)
- Understanding variable relationships
- Detecting multicollinearity

```python
# Create datasets with different correlations
np.random.seed(42)

# 1. Perfect positive correlation
x1 = np.array([1, 2, 3, 4, 5])
y1 = x1 * 2  # y = 2x

# 2. Positive correlation with noise
x2 = np.array([2, 3, 4, 5, 6, 7, 8])
y2 = x2 * 1.5 + np.random.normal(0, 2, len(x2))

# 3. No correlation
x3 = np.array([1, 2, 3, 4, 5])
y3 = np.array([3, 1, 4, 1, 5])  # Random values

print("CORRELATION ANALYSIS")
print("=" * 70)
print()

# Calculate correlations
corr1 = np.corrcoef(x1, y1)[0, 1]
corr2 = np.corrcoef(x2, y2)[0, 1]
corr3 = np.corrcoef(x3, y3)[0, 1]

print("1. PERFECT POSITIVE CORRELATION")
print("-" * 70)
print(f"x: {x1}")
print(f"y: {y1}")
print(f"Correlation: {corr1:.4f}")
print(f"Interpretation: When x increases, y ALWAYS increases proportionally")
print()

print("2. POSITIVE CORRELATION WITH NOISE")
print("-" * 70)
print(f"x: {x2}")
print(f"y: {np.round(y2, 2)}")
print(f"Correlation: {corr2:.4f}")
print(f"Interpretation: General upward trend, but with some variation")
print()

print("3. NO CORRELATION")
print("-" * 70)
print(f"x: {x3}")
print(f"y: {y3}")
print(f"Correlation: {corr3:.4f}")
print(f"Interpretation: x and y have no relationship")
print()

# Visualization
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Plot 1
axes[0].scatter(x1, y1, s=100)
axes[0].set_title(f'Perfect Positive (r={corr1:.3f})')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].grid(True, alpha=0.3)

# Plot 2
axes[1].scatter(x2, y2, s=100)
axes[1].set_title(f'Positive with Noise (r={corr2:.3f})')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].grid(True, alpha=0.3)

# Plot 3
axes[2].scatter(x3, y3, s=100)
axes[2].set_title(f'No Correlation (r={corr3:.3f})')
axes[2].set_xlabel('x')
axes[2].set_ylabel('y')
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("HOW TO USE CORRELATION IN ML")
print("=" * 70)
print("✓ Drop features not correlated with target")
print("✓ Keep features highly correlated with target")
print("✓ Remove redundant features (highly correlated with each other)")
print("✓ This improves model accuracy and reduces training time!")
```

## Section 2.4: Real-World Example - Feature Selection

### Problem
We have a dataset with 5 features predicting house prices. Which features should we use?

```python
# Create sample house data
np.random.seed(42)
n_samples = 100

# Features
size = np.random.uniform(1000, 5000, n_samples)
bedrooms = np.random.uniform(1, 6, n_samples)
age = np.random.uniform(0, 100, n_samples)
color = np.random.randint(0, 10, n_samples)  # Arbitrary
proximity_to_school = np.random.uniform(0, 10, n_samples)

# Target: Price (highly related to size, less to others)
price = size * 150 + bedrooms * 20000 - age * 500 + np.random.normal(0, 10000, n_samples)

print("FEATURE SELECTION USING CORRELATION")
print("=" * 70)
print()

# Calculate correlations
features = {
    'Size': size,
    'Bedrooms': bedrooms,
    'Age': age,
    'Color': color,
    'School Proximity': proximity_to_school
}

correlations = {}
for feature_name, feature_data in features.items():
    corr = np.corrcoef(feature_data, price)[0, 1]
    correlations[feature_name] = corr

# Sort by absolute correlation
sorted_corr = sorted(correlations.items(), key=lambda x: abs(x[1]), reverse=True)

print("Correlations with House Price:")
print("-" * 70)
for feature_name, corr in sorted_corr:
    strength = 'STRONG' if abs(corr) > 0.7 else 'MODERATE' if abs(corr) > 0.4 else 'WEAK'
    direction = 'positive' if corr > 0 else 'negative'
    print(f"{feature_name:20s}: {corr:7.4f} ({strength} {direction})")

print()
print("DECISION")
print("=" * 70)
print("✓ KEEP: Size (r≈0.99) - Very strong correlation")
print("✓ KEEP: Bedrooms (r≈0.80) - Strong correlation")
✓ KEEP: Age (r≈-0.48) - Moderate negative correlation")
✗ DROP: Color (r≈0.05) - No correlation with price")
✗ DROP: School Proximity (r≈0.12) - Weak correlation")
print()
print("RESULT")
print("-" * 70)
print("Model uses 3 features instead of 5")
print("Benefits:")
print("  • Faster training")
print("  • Better accuracy (removed noise)")
print("  • More interpretable model")
print("  • Less data needed")
```

## Section 2.5: Model Evaluation Metrics

### Key Metrics
- **MSE (Mean Squared Error):** Average squared difference from actual
- **RMSE (Root Mean Squared Error):** Square root of MSE (same units as target)
- **MAE (Mean Absolute Error):** Average absolute difference
- **R² Score:** Proportion of variance explained (0 to 1)

```python
# Actual vs predicted values
actual = np.array([100, 150, 200, 250, 300])
predicted = np.array([95, 155, 190, 260, 310])

print("MODEL EVALUATION METRICS")
print("=" * 70)
print()

print("Data:")
print("-" * 70)
for i, (a, p) in enumerate(zip(actual, predicted)):
    error = p - a
    print(f"Sample {i+1}: Actual={a}, Predicted={p}, Error={error}")
print()

# Calculate metrics
errors = predicted - actual
squared_errors = errors ** 2
abs_errors = np.abs(errors)

mse = np.mean(squared_errors)
rmse = np.sqrt(mse)
mae = np.mean(abs_errors)
ss_res = np.sum(squared_errors)
ss_tot = np.sum((actual - np.mean(actual)) ** 2)
r_squared = 1 - (ss_res / ss_tot)

print("Metrics:")
print("-" * 70)
print()
print(f"1. MSE (Mean Squared Error)")
print(f"   Formula: average(error²)")
print(f"   Errors squared: {squared_errors}")
print(f"   MSE = {mse:.2f}")
print(f"   Interpretation: Average squared error")
print()

print(f"2. RMSE (Root Mean Squared Error)")
print(f"   Formula: √MSE")
print(f"   RMSE = {rmse:.2f}")
print(f"   Interpretation: Predictions off by ±{rmse:.2f} on average")
print()

print(f"3. MAE (Mean Absolute Error)")
print(f"   Formula: average(|error|)")
print(f"   Absolute errors: {abs_errors}")
print(f"   MAE = {mae:.2f}")
print(f"   Interpretation: Average absolute error (less sensitive to outliers)")
print()

print(f"4. R² Score")
print(f"   Formula: 1 - (SS_res / SS_tot)")
print(f"   R² = {r_squared:.4f}")
print(f"   Interpretation: Model explains {r_squared*100:.2f}% of variance")
print(f"   Range: 0 (bad) to 1 (perfect)")
print()

print("WHICH METRIC TO USE?")
print("=" * 70)
print("✓ RMSE: When outliers should be penalized heavily")
print("✓ MAE: When all errors are equally important")
✓ R²: To understand model quality overall")
```
