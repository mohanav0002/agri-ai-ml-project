```python
# Input: current day after sowing
current_day = int(input("Enter current day after sowing: "))

# Growth stages with day ranges
stages = [
    ("Germination", 0, 7),
    ("Seedling", 8, 21),
    ("Vegetative", 22, 45),
    ("Flowering", 46, 65),
    ("Grain filling", 66, 85),
    ("Maturity", 86, float('inf'))
]

# Determine current stage
for stage, start, end in stages:
    if start <= current_day <= end:
        print(f"🌾 Current Growth Stage: {stage}")

        if end != float('inf'):
            days_remaining = end - current_day
            print(f"⏳ Days remaining in this stage: {days_remaining}")
        else:
            print("✅ Crop has reached maturity stage.")
        break
```
