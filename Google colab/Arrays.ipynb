```python
# Soil pH readings (3 fields × 4 samples)
ph_levels = np.array([
 [6.5, 6.8, 6.3, 6.7],
 [7.1, 7.0, 6.9, 7.2],
 [6.2, 6.4, 6.1, 6.3]
])
# Single element: row 1, column 2
print(ph_levels[1, 2])
```

```python
# Entire row 0
print(ph_levels[0, :])
```

```python
# Entire column 3
print(ph_levels[:, 3])
```

```python
# First 2 rows, first 3 columns
sub_array = ph_levels[0:2, 0:3]
print(sub_array)
```

```python

# Every other row, all columns
print(ph_levels[::2, :])
```

```python
temperatures = np.array([72, 85, 91, 68, 95, 70, 88])
# Create a boolean array (mask)
hot_days = temperatures > 85
print(hot_days)
```

```python
# Use mask to filter original array
print(temperatures[hot_days])
```

```python
# Can do it in one line
print(temperatures[temperatures > 85])
```

```python
# Combine conditions with & (and), | (or)
moderate = temperatures[(temperatures >= 70) &
(temperatures <= 85)]
print(moderate)
```

```python
# Find extreme values (very hot OR very cold)
extreme = temperatures[(temperatures > 90) |
(temperatures < 70)]
print(extreme)
```

```python
# Crop yields in tons
field_a = np.array([120, 135, 142, 128])
field_b = np.array([118, 140, 138, 132])
# Add arrays
total_yield = field_a + field_b
print(total_yield)
```

```python
# Subtract to find difference
difference = field_a - field_b
print(difference)
```

```python
# Multiply each by a constant
doubled = field_a * 2
print(doubled)
```

```python

prices = np.array([19.99, 24.51, 30.25])
# Round to nearest integer
np.round(prices)
# [20. 25. 30.]
# Floor (round down)
np.floor(prices)
# [19. 24. 30.]
# Ceiling (round up)
np.ceil(prices)
# [20. 25. 31.]
```
