```python
rows = 8
cols = 8

for r in range(rows):          # outer loop for rows
    for c in range(cols):      # inner loop for columns
        if (r + c) % 2 == 0:
            print("■", end=" ")   # dark square
        else:
            print("□", end=" ")   # light square
    print()  # new line after each row
```
